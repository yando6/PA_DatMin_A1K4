## Outlier

```python
def outlier(col):
    Q1 = x[col].quantile(0.25) # Mengetahui mana yang quartile 1
    Q3 = x[col].quantile(0.75) # Mengetahui mana yang quartile 3
    IQR = Q3 - Q1 # Simpangan Quartile (q3-q1)    
    return x[(x[col]> Q3 +1.5*IQR)  | (x[col]< Q1-1.5*IQR )]
```

## **K-Means**

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

Pada K-Means untuk dataset ini, kita akan mengelompokkan data berdasarkan umur, DesireToTakeInitiative dan Competitiveness

```python
x_2 = df.iloc[:,[2,7,8]].values
x_2[:5]
```

```python
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(x_2)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
plt.plot(range(1, 11), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Dalam elbow method, kita mencari jumlah cluster yang optimal dengan melihat elbow (siku) pada grafik yang menunjukkan penurunan yang signifikan dalam inertia. Pada contoh kode di atas, grafik elbow menunjukkan bahwa jumlah cluster optimal adalah 3.

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(x_2)


df['Cluster'] = y_kmeans + 1
df.head(5)
```

Setelah tahapan-tahapan tersebut dilakukan, kita dapat melihat hasil clustering dengan cara melihat tabel cluster pada dataset

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(111, projection= '3d')
ax.scatter(x_2[y_kmeans == 0,0],x_2[y_kmeans == 0,1],x_2[y_kmeans == 0,2], s = 40 , color = 'blue',label = 'cluster 1')
ax.scatter(x_2[y_kmeans == 1,0],x_2[y_kmeans == 1,1],x_2[y_kmeans == 1,2], s = 40 , color = 'orange',label = 'cluster 2')
ax.scatter(x_2[y_kmeans == 2,0],x_2[y_kmeans == 2,1],x_2[y_kmeans == 2,2], s = 40 , color = 'red',label = 'cluster 3')
ax.set_xlabel ( 'Usia-->')
ax.set_ylabel ( 'DesireToTakeInitiative-->')
ax.set_zlabel ( 'Competitiveness')
ax.legend()
plt.show()
```

Dalam plot tersebut, setiap titik merepresentasikan satu siswa, dengan sumbu x menunjukkan nilai Usia, sumbu y menunjukkan nilai DesireToTakeInitiative dan sumbu z menunjukkan nilai Competitiveness. Warna setiap titik menunjukkan cluster yang dimiliki oleh siswa tersebut.

Terdapat 3 Kelompok/cluster yang terbentuk yaitu

*   Cluster 1 : memiliki usia kisaran 18-21 tahun, memiliki tingkat keinginan untuk mengambil inisiatif kisaran 3 sampai 4,5 dan memiliki tingkat daya saing kisaran 2 sampai 3 poin

*   Cluster 2 : memiliki usia yang lumayan muda kisaran 17-19 tahun, memiliki tingkat keinginan untuk mengambil inisiatif yang sangat tinggi kisaran 4,5 sampai 5 dan memiliki tingkat daya saing kisaran 3 sampai 4 poin


*   Cluster 3 : memiliki usia yang lumayan muda kisaran 18-20 tahun, memiliki tingkat keinginan untuk mengambil inisiatif yang lumayan rendah kisaran 2 sampai 4 dan memiliki tingkat daya saing yang rendah kisaran 1 sampai 2,5 poin


```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Calculate silhouette score
silhouette = silhouette_score(x_2, kmeans.labels_)

# Calculate Calinski-Harabasz index
calinski = calinski_harabasz_score(x_2, kmeans.labels_)

# Calculate Davies-Bouldin index
davies_bouldin = davies_bouldin_score(x_2, kmeans.labels_)

# Create a DataFrame to store the scores
scores = pd.DataFrame({'Silhouette Score': [silhouette],
                       'Calinski-Harabasz Score': [calinski],
                       'Davies-Bouldin Score': [davies_bouldin]})

# Display the scores
print(scores)
```

Setelah melihat hasil evaluasi diatas maka dapat disimpulkan

> Silhouette Score yang diperoleh adalah 0.359559, yang menunjukkan bahwa pembagian cluster yang dihasilkan cukup baik karena nilai Silhouette Score yang lumayan.


> Calinski-Harabasz Score yang diperoleh adalah  123.428226, yang menunjukkan bahwa pembagian cluster yang dihasilkan cukup baik karena nilai Calinski-Harabasz Score yang tinggi.


> Davies-Bouldin Score yang diperoleh adalah 0.991683, yang menunjukkan bahwa pembagian cluster yang dihasilkan cukup baik karena nilai Davies-Bouldin Score yang rendah.


Secara keseluruhan,Silhouette Score, Calinski-Harabasz Score maupun Davies-Bouldin Score menunjukkan bahwa pembagian cluster yang dihasilkan cukup baik.
